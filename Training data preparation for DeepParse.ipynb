```python
import numpy as np
import pandas as pd
import pickle
from tqdm import tqdm
```

```python
raw_df = pd.read_csv('train.csv')
```

```python
df = raw_df.copy()
```

```python
df
```

```python
df['POI'] = df['POI/street'].apply(lambda x: x.split('/')[0].split())
df
```

```python
df['street'] = df['POI/street'].apply(lambda x: x.split('/')[1].split())
df
```

```python
df.head(20)
```

```python
target_column = list()
target = list()
for row in tqdm(df.itertuples(index = False), total = len(df)):
    for word in row[1].split():
        if word.strip(',') in row[4]:
            target.append('StreetName')
        elif word.strip(',') in row[3]:
            target.append('Unit')
        else:
            target.append('GeneralDelivery')
    target_column.append(target)
    target = list()
```

```python
target_column
```

```python
df['target'] = target_column
```

```python
df
```

## Verify correctness (the number of list elements and number of words)

```python
df['len_target'] = df['target'].apply(lambda x: len(x))
df['len_raw'] = df['raw_address'].apply(lambda x: len(x.split()))
```

```python
df
```

```python
np.sum(df['len_target'] != df['len_raw'])
```

## Save and export data

```python
df[['raw_address','target']].to_csv('deepparse_retrain.csv', index = False)
```
