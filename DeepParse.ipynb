## Retrain models

```python
training_container = PickleDatasetContainer('deepparse_retrain.p')
```

```python
training_container.data
```

```python
address_parser = AddressParser(model_type="bpemb", device=0)
```

```python
lr_scheduler = poutyne.StepLR(step_size=1, gamma=0.1)
```

```python
address_parser.retrain(training_container, 0.8, epochs=15, batch_size=64, num_workers=2, callbacks=[lr_scheduler])
```

```python
saving_dir = "./data"
file_extension = "p"
training_dataset_name = "sample_noisy_data"
test_dataset_name = "test_sample_data"
download_from_url(training_dataset_name, saving_dir, file_extension=file_extension)
download_from_url(test_dataset_name, saving_dir, file_extension=file_extension)
```

```python
sample_training_container = PickleDatasetContainer(os.path.join(saving_dir, training_dataset_name + "." + file_extension))
sample_test_container = PickleDatasetContainer(os.path.join(saving_dir, test_dataset_name + "." + file_extension))
```
